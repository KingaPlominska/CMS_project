Zadanie 1

```python
import numpy as np
import matplotlib.pyplot as plt #rysowanie wykresow

#Definiujemy czestotliwosc sygnalu sinusoidalnego
f=5

#Czestotliwosc probkowania
fs=1000 #HZ

#Definjujemy pocztaek i koniec
a=0
b=2

print(f'f = {f}') # formatuj tekst - wyswietl wartosc zmiennej

print(f'fs = {fs}')

print(f'a = {a}')

print(f'b = {b}')
```

```python
# utworz wektor czasu od 0 do 2sekun
t = np.arange(a, b, 1/fs)

print(f't={t}')
```

```python
# Utworz sygnal sinusoidy
sigl =np.sin(2*np.pi * f * t)

print(f's={sigl}')
```

```python
#Wyswietlanie
plt.plot(t,sigl)
plt.grid(alpha=0.2)
plt.title('sygnal sinusoidalny 5Hz')
plt.xlabel('czas [s]')
plt.ylabel('amplituda')
plt.show()
```

Zadanie 2


```python
#rozklad normlany, srednia = 0, ochylenie standardowe = 1
noise =np.random.normal (0,1, size=sigl.size)
```

```python
#wyswietlenie
plt.plot(t, noise)
plt.grid(alpha=0.2)
plt.title('sygnal szumowy')
plt.xlabel('czas [s]')
plt.ylabel('amplituda')
plt.show()
```

```python
#Mnozenie szumu (noise) przez 0.1
noise *= 0.1
```

```python
#Zaszumiony sygnal
sigNoisy = sigl + noise
```

```python
#Wyswietlenie
plt.plot(t,sigNoisy)
plt.grid(alpha=0.2)
plt.title('sygnal sinusoidalny zaszumiony 2Hz')
plt.xlabel('czas [s]')
plt.ylabel('amplituda')
plt.show()
```

```python
#Historia wartosci szumu
plt.hist(noise)
plt.show()
```

```python
from scipy import signal#Definicja czestotliwosci sygnalu
f=2

#stworz sygnal prostokatny
sq=signal.square(2* np.pi * f * t)
```

```python
# Wyswietlanie
plt.plot(t, sq)
plt.grid(alpha=0.2)
plt.title('sygnal prostokatny 2HZ')
plt.xlabel('czas [s]')
plt.ylabel('amplituda')
plt.show()
```

Zadanie 4


```python
#Zaszumiony sygnal
sigNoisy = sq + noise

#Wyswietlenie
plt.plot(t, sigNoisy)
plt.grid(alpha=0.2)
plt.title('sygnal prostokatny zaszumiony 2HZ')
plt.xlabel('czas [s]')
plt.ylabel('amplituda')
plt.show()
```

Zadanie 6


```python
#zmiana czestpotliwosci probkowania
fs = 100
t = np.arange(a, b, 1/fs)

#sygnal 1HZ
f1= 1
amplituda1 = 1
sig_1 = np.sin(2 * np.pi * f1 * t) * amplituda1

#sygnal 3HZ

f2= 3
amplituda2 = 1/3
sig_2 = np.sin(2 * np.pi * f2 * t) * amplituda2

#sygnal 5HZ

f3 = 5
amplituda3 = 1/5
sig_3 = np.sin(2* np.pi * f3 *t) * amplituda3

#Sygnal sumacyjny

sum_sig = sig_1 + sig_2 + sig_3

#Wyswietlanie
plt.plot(t, sum_sig)
plt.grid(alpha=0.2)
plt.title('suma sygnalow')
plt.xlabel('czas [s]')
plt.ylabel('amplituda')
plt.show()
```

Zadanie 7


```python
# Parametry sinusoid
frequencies = [1,3,4,7,9] #Czestotliwosc
amplitudes = [5,4,3,2,1] # Amplitudy

sum_signal = np.zeros(len(t))

for f, a in zip(frequencies, amplitudes):
  sig = np.sin(2 * np.pi * f * t) * a
  sum_signal += sig

#Wyswietlenie
plt.plot(t, sum_signal)
plt.grid(alpha=0.2)
plt.title('suma sygnalow')
plt.xlabel('czas [s]')
plt.ylabel('amplituda')
plt.show()
```
